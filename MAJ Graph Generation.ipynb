```python
import matplotlib.pyplot as plt
import networkx as nx
```

We used matplotlib and networkx to generate the following 4 graph types. 

# Lattice graph

Generated using networkx, with 10 rows and 10 columns (100 nodes). Hexagonal graphs means that each node has three connections, so it forms a hexagonal space.

```python
lattice = nx.hexagonal_lattice_graph(10, 10)
nx.draw(lattice, node_color = 'blue', width = .5, node_size= 50)
plt.savefig('lattice.png', format = 'PNG')
```

# Random graph

Generated using networkx, with 10 nodes and the probability of edge creation is 10.

```python
random = nx.gnp_random_graph(20,.5)
nx.draw(random, node_color = 'blue', width = .5, node_size= 100)
plt.savefig('random.png', format = 'PNG')
```

# Scale_free graph

Generated using networkx, with 10 nodes in the graph. This generates a graph with low heterogeneity (high clustering). 

```python
scale_free = nx.scale_free_graph(20)
nx.draw(scale_free, node_color = 'blue', width = .5, node_size= 50)
plt.savefig('scale_free.png', format = 'PNG')
```

# Small_world graph

Generated using networkx. In this algorithm n is the length of one side of the graph. Our number of nodes is n^2, or 100.  A navigable small-world graph is a directed grid with additional long-range
connections that are chosen randomly.

```python
small_world = nx.navigable_small_world_graph(10)
nx.draw_circular(small_world, node_color = 'blue', width = .5, node_size= 50)
plt.savefig('small_world.png', format = 'PNG')
```
