### Lattice model

```python
import networkx as net
import matplotlib.pyplot as plt

#Lettice
g = net.hexagonal_lattice_graph(3, 2)
net.draw(g, with_labels=True)
plt.title('Lettice')
plt.show();
```

### Random model

```python
#Random
g = net.erdos_renyi_graph(30, 0.1)
net.draw(g, with_labels=True)
plt.title('Random: Erods Renyi')
plt.show();
```

### SF model

```python
#Scale Free
g = net.barabasi_albert_graph(20, 10)
net.draw(g, with_labels=True)
plt.title('Scale Free')
plt.show
```

### SW model

```python
import networkx as net
import matplotlib.pyplot as plt
```

```python
G=nx.newman_watts_strogatz_graph(50, 3, 0.4)
```

```python
plt.figure(figsize=(7, 7))
net.draw_circular(G, with_labels=True, font_weight='bold', font_color='white', node_size=1000, node_color='purple', edge_color='gold',
                 style='dashdot', width=5)
plt.title('Small World graph / Newman-Watts Strogratz graph', fontsize='30')
plt.show()
```
