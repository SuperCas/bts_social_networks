We will use Python networkx generators to plot various types of social networks with differing distributions and levels of interconnectedness.

```python
import networkx as net
from networkx.generators import lattice, watts_strogatz_graph, scale_free_graph, erdos_renyi_graph
import matplotlib.pyplot as plt

#edges = traversal.bfs_edges(g, 0)
#list(edges)
```

Erdos-Renyi Graph (random). At low probabilities of connection, you can have unconnected nodes.

```python
g = erdos_renyi_graph(15,.2)
net.draw_circular(g, with_labels = True)
plt.show()
```

Lattice graph. Pretty homogeneous connections, except for nodes at the edge of network.

```python
g = lattice.grid_graph(dim = [5,4])
net.draw_circular(g, with_labels = True)
plt.show()
```

Small World Graph (Watts-Strogatz).  Nodes 8,14,16 and 6 are more highly connected, but other all nodes are pretty connected, low heterogeneity.

```python
g = watts_strogatz_graph(20,4,.2)
net.draw_circular(g, with_labels = True)
plt.show()
```

Scale-free: Nodes 1 and 0 are highly connected, most other nodes only connect to them. 

```python
g = scale_free_graph(20)
net.draw_circular(g, with_labels = True)
plt.show()
```
