We use Python networkx for graph generation.

```python
import networkx.generators.lattice as ngl
```

```python
import networkx as net
import matplotlib.pyplot as plt
%matplotlib inline
```

## Lattice Graph

```python
g1 = ngl.hexagonal_lattice_graph(5,5)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
net.draw(g1, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=400, width=2,
         font_color='white', font_weight='bold',
         ax=ax)
ax.set_title('Hexagonal Lattice Graph')
plt.show();
```

## Random Graphs

```python
import networkx.generators.random_graphs as rg
```

```python
g2_1=rg.fast_gnp_random_graph(20, 0.8)
g2_2=rg.fast_gnp_random_graph(20, 0.3)
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(16,8))

net.draw_circular(g2_1, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=100, width=2,
         font_color='white', font_weight='bold',
         ax=ax[0])
net.draw_circular(g2_2, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=100, width=2,
         font_color='white', font_weight='bold',
         ax=ax[1])
ax[0].set_title('Random Graph - high probability')
ax[1].set_title('Random Graph - low probability')
plt.show();
```

## Small World Graph

```python
g3_1=rg.watts_strogatz_graph(100,2,0.3)
g3_2=rg.watts_strogatz_graph(100,2,0.8)
g3_3=rg.watts_strogatz_graph(100,6,0.3)
g3_4=rg.watts_strogatz_graph(100,6,0.8)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2,figsize=(16,16))
net.draw_circular(g3_1, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=100, width=2,
         font_color='white', font_weight='bold',
         ax=ax[0,0])
net.draw_circular(g3_2, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=100, width=2,
         font_color='white', font_weight='bold',
         ax=ax[0,1])
net.draw_circular(g3_3, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=100, width=2,
         font_color='white', font_weight='bold',
         ax=ax[1,0])
net.draw_circular(g3_4, with_labels = False, node_color='indigo', 
         edge_color='indigo', node_size=100, width=2,
         font_color='white', font_weight='bold',
         ax=ax[1,1])
ax[0,0].set_title('Small World Graph - few neighbors, low probability')
ax[0,1].set_title('Random Graph - few neighbors, high probability')
ax[1,0].set_title('Random Graph - many neighbors, low probability')
ax[1,1].set_title('Random Graph - many neighbors, high probability')
plt.show();
```
